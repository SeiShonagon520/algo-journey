```python
files = ["data.txt", "photo.jpg", "script.py", "image.png", "report.doc", "avatar.jpg"]
result = [i for i in files if i.endswith((".jpg",".png"))]
print(result)
```

```python
scores = [55, 80, 92, 45, 70]
result = [i if i >= 60 else "Fail" for i in scores]
print(result)
```
