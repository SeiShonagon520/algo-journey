## 2. Python ä¸­çš„æ ˆ

åœ¨ Python ä¸­ï¼Œæˆ‘ä»¬ä¸éœ€è¦å¼•å…¥é¢å¤–çš„åº“ï¼Œç›´æ¥ä½¿ç”¨å†…ç½®çš„ **åˆ—è¡¨ (`list`)** å°±å¯ä»¥å®Œç¾å®ç°æ ˆçš„æ‰€æœ‰åŠŸèƒ½ã€‚

| :--- | :--- | :--- | :--- |
| **Push** | å…¥æ ˆï¼ˆåŠ åˆ°é¡¶ç«¯ï¼‰ | `stack.append(x)` | $O(1)$ |
| **Pop** | å‡ºæ ˆï¼ˆæ‹¿èµ°é¡¶ç«¯ï¼‰ | `stack.pop()` | $O(1)$ |

## 4. å®æˆ˜æ¡ˆä¾‹ï¼šæœ‰æ•ˆçš„æ‹¬å·

**é¢˜ç›®**ï¼šç»™å®šä¸€ä¸ªåªåŒ…æ‹¬ `'('`, `')'`, `'{'`, `'}'`, `'['`, `']'` çš„å­—ç¬¦ä¸²ï¼Œåˆ¤æ–­å­—ç¬¦ä¸²æ˜¯å¦æœ‰æ•ˆã€‚

**æ€è·¯**ï¼š
1.  é‡åˆ°**å·¦æ‹¬å·**ï¼šå‹å…¥æ ˆä¸­ï¼ˆå› ä¸ºå®ƒåœ¨ç­‰å¾…ä¸€ä¸ªå³æ‹¬å·æ¥åŒ¹é…ï¼‰ã€‚
2.  é‡åˆ°**å³æ‹¬å·**ï¼š
    *   æ£€æŸ¥æ ˆæ˜¯å¦ä¸ºç©ºï¼Ÿï¼ˆç©ºäº†è¯´æ˜æ²¡æœ‰å·¦æ‹¬å·èƒ½åŒ¹é…å®ƒ -> éæ³•ï¼‰
    *   æ£€æŸ¥æ ˆé¡¶çš„å·¦æ‹¬å·å’Œå½“å‰å³æ‹¬å·æ˜¯ä¸€å¯¹å—ï¼Ÿï¼ˆ`(` é… `)`ï¼Œ`[` é… `]`...ï¼‰
        *   æ˜¯ï¼šåŒ¹é…æˆåŠŸï¼Œæ ˆé¡¶å·¦æ‹¬å·å¼¹å‡ºï¼ˆé”€è´¦ï¼‰ã€‚
        *   å¦ï¼šç±»å‹ä¸åŒ¹é… -> éæ³•ã€‚
3.  éå†ç»“æŸåï¼š
    *   å¦‚æœæ ˆç©ºäº†ï¼šè¯´æ˜æ‰€æœ‰å·¦æ‹¬å·éƒ½æ‰¾åˆ°äº†å½’å®¿ -> åˆæ³•ã€‚
    *   å¦‚æœæ ˆä¸ç©ºï¼šè¯´æ˜è¿˜æœ‰å·¦æ‹¬å·æ²¡ç­‰åˆ°å®ƒçš„å¦ä¸€åŠ -> éæ³•ã€‚

```python
def isValid(s: str) -> bool:
    # å­—å…¸ç”¨æ¥å­˜åŒ¹é…å…³ç³»ï¼Œå³æ‹¬å·å½“ Key æŸ¥èµ·æ¥æ›´æ–¹ä¾¿
    pairs = {
        ')': '(',
        ']': '[',
        '}': '{'
    }
    stack = []
    
    for char in s:
        if char in pairs: 
            # === é‡åˆ°å³æ‹¬å· ===
            # å¦‚æœæ ˆç©ºï¼Œæˆ–è€…æ ˆé¡¶å…ƒç´ ä¸æ˜¯å¯¹åº”çš„å·¦æ‹¬å·
            if not stack or stack[-1] != pairs[char]:
                return False
            stack.pop() # åŒ¹é…æˆåŠŸï¼Œæ¶ˆæ¶ˆä¹
        else:
            # === é‡åˆ°å·¦æ‹¬å· ===
            stack.append(char)
            
    # æœ€åå¿…é¡»æ ˆç©ºæ‰ç®—å®Œå…¨åŒ¹é…
    return len(stack) == 0

# æµ‹è¯•
print(isValid("()[]{}"))  # True
print(isValid("(]"))      # False
print(isValid("([)]"))    # False
print(isValid("{[]}"))    # True
```
