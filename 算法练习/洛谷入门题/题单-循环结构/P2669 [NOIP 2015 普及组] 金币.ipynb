# P2669 [NOIP 2015 普及组] 金币

## 题目背景

NOIP2015 普及组 T1

## 题目描述

国王将金币作为工资，发放给忠诚的骑士。第一天，骑士收到一枚金币；之后两天（第二天和第三天），每天收到两枚金币；之后三天（第四、五、六天），每天收到三枚金币；之后四天（第七、八、九、十天），每天收到四枚金币……；这种工资发放模式会一直这样延续下去：当连续 $n$ 天每天收到 $n$ 枚金币后，骑士会在之后的连续 $n+1$ 天里，每天收到 $n+1$ 枚金币。

请计算在前 $k$ 天里，骑士一共获得了多少金币。

## 输入格式

一个正整数 $k$，表示发放金币的天数。

## 输出格式

一个正整数，即骑士收到的金币数。

## 输入输出样例 #1

### 输入 #1

```
6
```

### 输出 #1

```
14
```

## 输入输出样例 #2

### 输入 #2

```
1000
```

### 输出 #2

```
29820
```

## 说明/提示

**【样例 1 说明】**

骑士第一天收到一枚金币；第二天和第三天，每天收到两枚金币；第四、五、六天，每天收到三枚金币。因此一共收到 $1+2+2+3+3+3=14$ 枚金币。


对于 $100\%$ 的数据，$1\le k\le 10^4$。

````python
#输入发放金币的总天数
k_day = int(input())
#设置一个开关，用于结束外层循环
flag = False
#初始化金币数和天数和已过的天数
gold = 0
day = 0
to_day = 0
#遍历，外层循环，i表示每天发的金币数量
for i in range(1,k_day + 1):
    #开关
    if flag:
        break
    #遍历，内层循环，day当前的天数
    for day in range (1,i + 1):
        #判断经过的天数是否大于总天数
        if to_day >= k_day:
            #开启，结束外层循环和内层循环
            flag = True
            break
        #金币数和已过天数增加
        gold += i
        to_day += 1

print(gold)
````
