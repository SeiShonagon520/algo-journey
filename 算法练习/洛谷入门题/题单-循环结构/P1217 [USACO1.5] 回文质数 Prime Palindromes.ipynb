# P1217 [USACO1.5] 回文质数 Prime Palindromes

## 题目描述

因为 $151$ 既是一个质数又是一个回文数（从左到右和从右到左是看一样的），所以 $151$ 是回文质数。

写一个程序来找出范围 $[a,b] (5 \le a < b \le 100,000,000)$（一亿）间的所有回文质数。

## 输入格式

第一行输入两个正整数 $a$ 和 $b$。

## 输出格式

输出一个回文质数的列表，一行一个。

## 输入输出样例 #1

### 输入 #1

```
5 500
```

### 输出 #1

```
5
7
11
101
131
151
181
191
313
353
373
383
```

## 说明/提示

Hint 1: Generate the palindromes and see if they are prime.

提示 1: 找出所有的回文数再判断它们是不是质数（素数）.


Hint 2: Generate palindromes by combining digits properly. You might need more than one of the loops like below.

````python
a, b = map(int, input().split())

# 判断质数
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# 检查并输出
def check(p):
    if p < a:
        return True   # 继续
    if p > b:
        return False  # 停止
    if is_prime(p):
        print(p)
    return True

# 1 位回文数
for d1 in range(1, 10):
    p = d1
    if not check(p):
        break

# 2 位回文数 (只有 11 是质数)
for d1 in range(1, 10):
    p = d1 * 10 + d1
    if not check(p):
        break

# 3 位回文数: d1 d2 d1
for d1 in range(1, 10):
    for d2 in range(0, 10):
        p = d1 * 100 + d2 * 10 + d1
        if not check(p):
            break

# 5 位回文数: d1 d2 d3 d2 d1
for d1 in range(1, 10):
    for d2 in range(0, 10):
        for d3 in range(0, 10):
            p = d1 * 10000 + d2 * 1000 + d3 * 100 + d2 * 10 + d1
            if not check(p):
                break

# 7 位回文数: d1 d2 d3 d4 d3 d2 d1
for d1 in range(1, 10):
    for d2 in range(0, 10):
        for d3 in range(0, 10):
            for d4 in range(0, 10):
                p = d1 * 1000000 + d2 * 100000 + d3 * 10000 + d4 * 1000 + d3 * 100 + d2 * 10 + d1
                if not check(p):
                    break
````
