递归

```python
def recur(n:int):
    if n == 1:#终止条件
        return 1
    res = recur(n-1)
    return res + n

if __name__ == '__main__':
    print(recur(5))
```

尾递归

```python
def recur(n,res):
    if n == 0:
        return res
    return recur(n-1,res+n)

if __name__ == '__main__':
    print(recur(5,0))
```

递归树

```python
#斐波那契数列
def fib(n:int):
    if n == 1 or n == 2:
        return n -1
    res = fib(n-1) + fib(n-2)
    return res

if __name__ == '__main__':
    n = 5
    res = fib(n)
    print(res)
```

```python
#阶乘的递归
def fac(n:int):
    if n == 0:
        return 1
    res = n * fac(n-1)
    return res

if __name__ == '__main__':
    n = 5
    res = fac(n)
    print(res)
```
