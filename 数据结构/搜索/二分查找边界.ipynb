### 复用查找左边界

```python
def binary_search_right_edge(nums,target):
    i,j = 0,len(nums) -1
    ans = len(nums)
    while i <= j:
        m = (i+j)//2
        if nums[m] > target:
            ans = m
            j = m -1
        else:
            i = m + 1
    return ans - 1

a = [1,2,3,4,5,5,5,6,8,9]
result = binary_search_right_edge(a, 5)
print(result)
```
