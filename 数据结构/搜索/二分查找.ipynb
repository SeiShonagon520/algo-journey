# 二分查找


二分查找（binary search）是一种基于分治策略的高效搜索算法。它利用数据的有序性，每轮缩小一半搜索范围，直至找到目标元素或搜索区间为空为止

```python
def binary_search(nums,target):
    i, j = 0,len(nums) - 1
    while i <= j:
        m = (i + j) // 2
        if nums[m] < target:
            i = m + 1
        elif nums[m] > target:
            j = m - 1
        else:
            return m
    return -1
```

```python
a = [1,5,6,7,8,9,11,25]
binary_search(a,8)
```
