# 二分查找插入点

二分查找不仅可用于搜索目标元素，还可用于解决许多变种问题，比如搜索目标元素的插入位置。

## 存在重复元素的情况

```python
def binary_search_insertion(nums,target):
    i, j = 0, len(nums) - 1
    while i <= j:
        m = (i + j) // 2
        if nums[m] < target:
            i = m + 1
        elif nums[m] > target:
            j = m - 1
        else:
            j = m -1
    return i
```

```python
a = [1,3,5,5,5,6,9]
binary_search_insertion(a,5)
```
