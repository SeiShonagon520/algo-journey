哈希表（hash table），又称散列表，它通过建立键 key 与值 value 之间的映射，实现高效的元素查询。

![image](../images/hash_table_lookup.png)

```python
if __name__ == "__main__":
    hmap = {}
    hmap[110] = "小睿"
    hmap[111] = "小王"
    
    # 打印特定值
    name = hmap[110]
    print(f"学号110的学生是: {name}")
    
    # 打印所有值
    print("所有学生姓名:")
    for value in hmap.values():
        print(value)
    
    # 打印所有键值对
    print("完整信息:")
    for key, value in hmap.items():
        print(f"学号: {key}, 姓名: {value}")
```
