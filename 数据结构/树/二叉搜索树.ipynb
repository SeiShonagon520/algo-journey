二叉搜索树 (BST) 的核心规则
**左小右大：**

    1.左子树所有节点的值都小于根节点。

      右子树所有节点的值都大于根节点。

    2.任意节点的左、右子树也是二叉搜索树，即同样满足条件 1.

![](binary_search_tree.png)

查找节点
给定目标节点值 num ，可以根据二叉搜索树的性质来查找。如图 7-17 所示，我们声明一个节点 cur ，从二叉树的根节点 root 出发，循环比较节点值 cur.val 和 num 之间的大小关系。

若 cur.val < num ，说明目标节点在 cur 的右子树中，因此执行 cur = cur.right 。
若 cur.val > num ，说明目标节点在 cur 的左子树中，因此执行 cur = cur.left 。
若 cur.val = num ，说明找到目标节点，跳出循环并返回该节点。

查找结点

```python
class TreeNode:
    """
    定义二叉树的节点。
    """
    def __init__(self, key):
        self.val = key
        self.left = None
        self.right = None

def insert(root, key):
    """
    在二叉搜索树中插入一个新节点。
    """
    if root is None:
        return TreeNode(key)
    else:
        if root.val < key:
            root.right = insert(root.right, key)
        else:
            root.left = insert(root.left, key)
    return root

def search(root, key):
    cur =  root
    while cur is not None:

        if cur.val <key:
            cur = cur.right

        elif cur.val > key:
            cur = cur.left

        else:
            break

    return  cur
# 以这个顺序插入，可以确保生成和图片一模一样的树
values_to_insert = [8, 4, 12, 2, 6, 10, 14, 1, 3, 5, 7, 9, 11, 13, 15]

# 初始化一个空树
root = None

# 遍历列表，将每个值插入树中
for value in values_to_insert:
    root = insert(root, value)

node = search(root, 55)
if node is not None:
    print(node.val)
else:
    print("没有找到目标值。")
```

插入结点

```python
class TreeNode:
    """
    定义二叉树的节点。
    """
    def __init__(self, key):
        self.val = key
        self.left = None
        self.right = None

def insert(root, num: int):
    """插入节点"""
    # 若树为空，则初始化根节点
    if root is None:
        return TreeNode(num)

    # 循环查找，越过叶节点后跳出
    cur, pre = root, None
    while cur is not None:
        # 找到重复节点，直接返回
        if cur.val == num:
            return
        pre = cur
        # 插入位置在 cur 的右子树中
        if cur.val < num:
            cur = cur.right
        # 插入位置在 cur 的左子树中
        else:
            cur = cur.left
    # 插入节点
    node = TreeNode(num)
    if pre.val < num:
        pre.right = node
    else:
        pre.left = node

    return root


def search(root, key):
    cur =  root
    while cur is not None:

        if cur.val <key:
            cur = cur.right

        elif cur.val > key:
            cur = cur.left

        else:
            break

    return  cur
# 以这个顺序插入，可以确保生成和图片一模一样的树
values_to_insert = [8, 4, 12, 2, 6, 10, 14, 1, 3, 5, 7, 9, 11, 13, 15]

# 初始化一个空树
root = None

# 遍历列表，将每个值插入树中
for value in values_to_insert:
    root = insert(root, value)

node = search(root, 55)
if node is not None:
    print(node.val)
else:
    print("没有找到目标值。")

root = insert(root, 16)

node = search(root, 16)
if node is not None:
    print(node.val)
else:
    print("没有找到目标值。")
```
