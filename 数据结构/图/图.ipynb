图是一种极其重要和灵活的数据结构，用于表示对象之间“多对多”的复杂关系。在社交网络、地图导航、计算机网络、任务调度等众多领域都有广泛应用。

图的基本概念 (Basic Concepts)
一个图 G 由两个集合组成：顶点（Vertices）集合 V 和边（Edges）集合 E。

顶点 (Vertex/Node)：图中的一个基本单元，可以代表任何实体，比如一个城市、一个人、一个任务或一个网页。

边 (Edge)：连接两个顶点的线，表示顶点之间的关系。例如，城市之间的道路、人与人之间的朋友关系等。一条边 e 连接顶点 u 和 v，通常表示为 (u, v)。

图数据结构包含以下常用术语。

邻接（adjacency）：当两顶点之间存在边相连时，称这两顶点“邻接”。在图 9-4 中，顶点 1 的邻接顶点为顶点 2、3、5。

路径（path）：从顶点 A 到顶点 B 经过的边构成的序列被称为从 A 到 B 的“路径”。在图 9-4 中，边序列 1-5-2-4 是顶点 1 到顶点 4 的一条路径。

度（degree）：一个顶点拥有的边数。对于有向图，入度（in-degree）表示有多少条边指向该顶点，出度（out-degree）表示有多少条边从该顶点指出。

# 初始化图

```python
def __init__(self,vertices:list[int],edges:list[list[int]]):
    self.vertices :list[int] = []
    self.adj_mat :list[list[int]] = []

    for val in vertices:
        self.add_vertex(val)
    for e in edges:
        self.add_edge(e[0],e[1])
```
