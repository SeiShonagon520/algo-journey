# 高效输入
## 1. map + split
### 
```python
a, b, c = map(int, input().split())
```

````python
a, b, c = map(int, input().split())#输入10 30 50

print(a,b,c)
````

技术拆解 (Under the hood)：

input(): 读入一整行字符串 "10 20 30"。

.split(): 按空格切割，变成字符串列表 ['10', '20', '30']。

map(int, ...): 将列表里的每一个字符串 '10' 映射（转换）为整数 10。

a, b, c = ...: 解包 (Unpacking)。将迭代器里的三个数分别赋值给变量。

### 单行输入转列表


````python
nums = list(map(int, input().split()))
print(nums)
````

### 多行输入（纵向读取）

````python
n = int(input()) #输入行数（适用于第一行告诉行数内容的情况）
# 循环n次，每读一行加入进列表
lst = [int(input()) for _ in range(n)]
````
