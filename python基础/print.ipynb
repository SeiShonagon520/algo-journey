# 格式化输出（f-string）

使用 `f"..."` 可以在字符串中嵌入变量和格式控制。

```python
# 基本用法：嵌入变量
name = "小明"
age = 18
print(f"我叫{name}，今年{age}岁")  # 我叫小明，今年18岁

# 控制小数位数
x = 8
print(f"{x:.2f}")   # 8.00  保留2位小数
print(f"{x:.1f}")   # 8.0   保留1位小数
print(f"{x:.3f}")   # 8.000 保留3位小数

# 实际应用：斐波那契结果
result = 8
print(f"{result:.2f}")  # 8.00
```

## 格式说明

| 写法 | 含义 | 例子 |
|------|------|------|
| `{x}` | 直接输出变量 | `8` |
| `{x:.2f}` | 保留2位小数 | `8.00` |
| `{x:.1f}` | 保留1位小数 | `8.0` |
| `{x:05d}` | 整数补零到5位 | `00008` |

**记忆技巧**：`.2f` = 点 + 位数 + f（float浮点数）
