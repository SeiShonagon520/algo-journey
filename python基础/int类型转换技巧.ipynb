# int() 类型转换技巧

## 自动去除前导零

`int()` 把字符串转成整数时，会**自动去掉前导零**。

这在数字反转等题目中非常有用！

```python
# 正数：自动去掉前导零
print(int("083"))      # 83
print(int("00123"))    # 123
print(int("007"))      # 7

# 负数：同样有效
print(int("-083"))     # -83
print(int("-00456"))   # -456
```

## 实际应用：数字反转

比如 `-380` 反转后应该是 `-83`，而不是 `-083`。

用 `int()` 可以一步到位！

```python
# 数字反转示例
n = "-380"

# 处理负号
n = n.replace('-', '')
res = '-' + n[::-1]    # res = "-083"

# 用 int() 自动去掉前导零
print(int(res))        # -83
```

## 注意事项

- `int()` 只能转换**纯数字字符串**（可以有负号）
- 如果字符串包含非数字字符会报错

```python
# 正确用法
print(int("123"))      # 123
print(int("-456"))     # -456

# 错误用法（会报错）
# int("12.3")   # 有小数点，报错
# int("abc")    # 非数字，报错
```
