# 无穷大 `float('inf')`

Python 中表示正无穷大和负无穷大的方法。

## 基本用法

```python
# 正无穷大
pos_inf = float('inf')

# 负无穷大
neg_inf = float('-inf')

print(pos_inf)  # inf
print(neg_inf)  # -inf
```

## 特性：比任何数都大/小

```python
x = float('inf')

print(x > 9999999999)   # True
print(x > 10**100)      # True，比任何数都大
print(100 < x)          # True

y = float('-inf')
print(y < -9999999999)  # True，比任何数都小
```

## 常见应用场景

| 场景 | 初始值 |
|------|--------|
| 求**最小值** | `float('inf')` |
| 求**最大值** | `float('-inf')` |

```python
# 示例：求列表中的最小值
nums = [5, 2, 8, 1, 9]

min_val = float('inf')  # 初始设为无穷大
for num in nums:
    if num < min_val:
        min_val = num

print(min_val)  # 1
```

```python
# 示例：求列表中的最大值
nums = [5, 2, 8, 1, 9]

max_val = float('-inf')  # 初始设为负无穷
for num in nums:
    if num > max_val:
        max_val = num

print(max_val)  # 9
```

## 优点

- **永远不会出错**：不用估算数据范围的最大值
- **代码意图清晰**：一看就知道在求最小/最大值
- **Python 惯用写法**：算法题中的标准做法
