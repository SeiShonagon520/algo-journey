# 字符串处理常用函数

## `replace()` 替换

把字符串中的某个内容替换成另一个内容。

```python
s = "0-670-82162-4"

print(s.replace('-', ''))      # 0670821624   删除所有 -
print(s.replace('-', ' '))     # 0 670 82162 4  把 - 换成空格
print(s.replace('0', 'X'))     # X-67X-82162-4  把 0 换成 X
```

## `split()` 分割

按指定分隔符把字符串拆成列表。

```python
# 默认按空格分割
print("a b c".split())         # ['a', 'b', 'c']

# 指定分隔符
print("a-b-c".split('-'))      # ['a', 'b', 'c']
print("a,b,c".split(','))      # ['a', 'b', 'c']
print("a::b::c".split('::'))   # ['a', 'b', 'c']
```

## 字符串索引

字符串可以像列表一样用下标访问每个字符。

```python
s = "hello"

print(s[0])   # 'h'  第1个字符
print(s[1])   # 'e'  第2个字符
print(s[-1])  # 'o'  最后1个字符
print(len(s)) # 5    字符串长度
```

## 遍历字符串

```python
s = "abc"

# 方法1：直接遍历字符
for char in s:
    print(char)  # 依次输出 a, b, c

# 方法2：用索引遍历
for i in range(len(s)):
    print(i, s[i])  # 0 a, 1 b, 2 c
```
