# 循环控制技巧

## `break` 只能跳出一层循环

在嵌套循环中，`break` 只会跳出**当前所在的那一层**循环，外层循环会继续执行。

```python
# break 只跳出内层循环
for i in range(3):
    for j in range(3):
        if j == 1:
            break      # 只跳出内层 for j
        print(i, j)

# 输出: 0 0, 1 0, 2 0（外层继续跑）
```

## 用标志变量跳出多层循环

用一个布尔变量当"信号"，内层设为 True，外层检测到就 break。

常见命名：`flag`、`done`、`found`

```python
flag = False                    # 初始：旗帜放下

for i in range(10):
    if flag:                    # 检查旗帜
        break                   # 跳出外层
    for j in range(10):
        if i == 2 and j == 3:   # 某个条件满足
            flag = True         # 举旗！
            break               # 跳出内层
        print(i, j)
```

## for vs while

- **for**: 知道循环几次（如遍历列表、`range(n)`）
- **while**: 不知道循环几次，满足条件就继续

```python
# for: 明确循环 5 次
for i in range(5):
    print(i)

# while: 不知道几次，直到条件不满足
n = 1
while n < 100:
    n = n * 2
print(n)  # 128
```
