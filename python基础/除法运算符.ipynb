# Python 除法运算符

## 一句话区分

| 运算符 | 名称 | 作用 | 记忆口诀 |
|--------|------|------|----------|
| `/` | 普通除法 | 得到小数结果 | 一杠带小数 |
| `//` | 整除 | 只要整数部分 | 两杠砍小数 |
| `%` | 取余 | 只要余数 | 百分号要余数 |

## 具体例子：17 除以 5

```python
# 普通除法：得到完整的小数结果
print(17 / 5)   # 输出 3.4
```

```python
# 整除：只要商的整数部分
print(17 // 5)  # 输出 3
```

```python
# 取余：只要除法剩下的余数
# 17 = 5 x 3 + 2，所以余数是 2
print(17 % 5)   # 输出 2
```

## 记忆技巧：分糖果

想象：**17 颗糖分给 5 个人**

- `/` 平均每人得到 **3.4** 颗（理论值）
- `//` 每人实际拿走 **3** 颗整糖
- `%` 分完后盒子里**剩 2** 颗

## 符号记忆

- `/` 一条杠 = 结果完整带小数
- `//` 两条杠 = 像刀把小数砍掉
- `%` 两个圈被分开 = 剩下的部分

## 常见用途

- `//` 整除：计算有多少组，如 `n // 3` 表示能分几组
- `%` 取余：判断奇偶，如 `n % 2 == 0` 判断是否偶数
